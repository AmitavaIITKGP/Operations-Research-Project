```python
import numpy as np
import pandas as pd
from fractions import Fraction # so that numbers are not displayed in decimal.

```

```python
data=pd.read_excel('fish_feed_formulation.xlsx')
df=pd.DataFrame(data)
df
```

```python

# A will contain the coefficients of the constraints
constraint=list()
percentage=list()
n=int(input("Enter number of nutrients to be constrained: "))
for i in range(0,n):
    constraint.append(input("Enter nutrient to be constrained: "))
    percentage.append(float(input("Enter percentage required of the nutrient")))
prod_wt=float(input("Enter weight of product to be produced"))
MIN=int(input("Enter 1 for minimization and 0 for maximisation"))
A = df[constraint].to_numpy().transpose()
A=A/100
no_decision_var=len(df)
A=np.vstack((np.ones(no_decision_var),A))

id=np.identity(n+1)
A=np.hstack((A,id))
print(pd.DataFrame(A))
```

```python
percentage=np.array(percentage)
nutrient_wt=(percentage/100)*prod_wt
b=np.hstack((prod_wt,nutrient_wt))
```

```python
c=df[["Price"]].to_numpy().transpose()
M=999999
z=np.ones([1,n+1])*(-M)
if(MIN==1):
    c=-c
c=np.hstack((c,z))




```

```python
cb=c[:,no_decision_var:].transpose()
cb
```

```python
B=np.arange(no_decision_var,no_decision_var+n+1).reshape(n+1,1)
```

```python

xb=b.reshape(n+1,1)
```

```python
# combine matrices B and cb
table = np.hstack((B, cb))			
table = np.hstack((table, xb))		
# combine matrices B, cb and xb
# finally combine matrix A to form the complete simplex table
table = np.hstack((table, A))		
# change the type of table to float
table = np.array(table, dtype ='float')
# inputs end
df1=pd.DataFrame(table)
df1
```

```python
import sys
# if min problem, make this var 1


print("Table at itr = 0")
print("B \tCB \tXB \ty1 \ty2 \ty3 \ty4")
print(pd.DataFrame(table))

print()
print("Simplex Working....")

# when optimality reached it will be made 1
reached = 0	
itr = 1
unbounded = 0
alternate = 0

while reached == 0:

	print("Iteration: ", end =' ')
	print(itr)
	print("B \tCB \tXB \ty1 \ty2 \ty3 \ty4")
	print(pd.DataFrame(table))
	print()

	# calculate Relative profits-> cj - zj for non-basics
	i = 0
	rel_prof = []
	while i<len(A[0]):
		rel_prof.append(c[0,i] - np.sum(table[:, 1]*table[:, 3 + i]))
		i = i + 1

	print("rel profit: ", end =" ")
	for profit in rel_prof:
		print(Fraction(str(profit)).limit_denominator(100), end =", ")
	print()
	i = 0
	
	b_var = table[:, 0]
	# checking for alternate solution
	while i<len(A[0]):
		j = 0
		present = 0
		while j<len(b_var):
			if int(b_var[j]) == i:
				present = 1
				break;
			j+= 1
		if present == 0:
			if rel_prof[i] == 0:
				alternate = 1
				print("Case of Alternate found")
				# print(i, end =" ")
		i+= 1
	print()
	flag = 0
	for profit in rel_prof:
		if profit>0:
			flag = 1
			break
		# if all relative profits <= 0
	if flag == 0:
		print("All profits are >= 0, optimality reached")
		reached = 1
		break

	# kth var will enter the basis
	k = rel_prof.index(max(rel_prof))
	min = 99999
	i = 0;
	r = -1
	# min ratio test (only positive values)
	while i<len(table):
		if (table[:, 2][i]>0 and table[:, 3 + k][i]>0):
			val = table[:, 2][i]/table[:, 3 + k][i]
			if val<min:
				min = val
				r = i	 # leaving variable
		i+= 1

		# if no min ratio test was performed
	if r ==-1:
		unbounded = 1
		print("Case of Unbounded")
		break

	print("pivot element index:", end =' ')
	print(np.array([r, 3 + k]))

	pivot = table[r][3 + k]
	print("pivot element: ", end =" ")
	print(Fraction(pivot).limit_denominator(100))
		
		# perform row operations
	# divide the pivot row with the pivot element
	table[r, 2:len(table[0])] = table[
			r, 2:len(table[0])] / pivot
			
	# do row operation on other rows
	i = 0
	while i<len(table):
		if i != r:
			table[i, 2:len(table[0])] = table[i,2:len(table[0])] - table[i][3 + k] * table[r, 2:len(table[0])]
				
				
		i += 1

	
	# assign the new basic variable
	table[r][0] = k
	table[r][1] = c[0,k]
	
	print()
	print()
	itr+= 1
	

print()

print("***************************************************************")

m=0
while m<len(table): #If feasible solution does not exist, prints warning and exits code
    if table[m][0]>=df.shape[0] :
        print("Feasible Solution Does not Exist! Kindly Change Input Parameters")
        print()
        sys.exit()
    m+=1
    
if unbounded == 1:
	print("UNBOUNDED LPP")
	exit()
if alternate == 1:
	print("ALTERNATE Solution")

print("optimal table:")
print("B \tCB \tXB \ty1 \ty2 \ty3 \ty4")
print(pd.DataFrame(table))
print()
print("value of Z at optimality: ", end =" ")



basis = []
i = 0
sum = 0
while i<len(table):
	sum += c[0,int(table[i][0])]*table[i][2]
	temp = "x"+str(int(table[i][0])+1)
	basis.append(temp)
	i+= 1
# if MIN problem make z negative
if MIN == 1:
	print(-sum)
else:
	print(sum)
print("Final Basis: ", end =" ")
print(basis)

print("Simplex Finished...")
print()

ingredients=df[["Ingredients"]].to_numpy()
ingredients

for i in range(0,len(table)):
    print("Amount of ",ingredients[i]," required is ",table[i][2]," kgs")
    
```
